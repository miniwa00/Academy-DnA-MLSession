```python
import pandas as pd
import numpy as np
```

```python
tr = pd.read_csv('train.csv', encoding = 'cp949')
te = pd.read_csv('test.csv', encoding = 'cp949')
```

```python
# 타겟값 형변환
tr['credit'] = tr['credit'].astype('str')
```

```python
# 결측치 처리
tr['occyp_type'] = tr['occyp_type'].fillna('no job')
te['occyp_type'] = te['occyp_type'].fillna('no job')
```

```python
# 값이 모두 같은 열 제거
tr.drop(columns = 'FLAG_MOBIL')
te.drop(columns = 'FLAG_MOBIL')
```

```python
tr.info()
```

```python
# income total

# 로그 변환
tr['income_total_log'] = tr['income_total'].apply(np.log1p)
te['income_total_log'] = te['income_total'].apply(np.log1p)
```

```python
#days_birth

# 출생일 -> 나이
tr['age'] = (abs(tr['DAYS_BIRTH'])/365).astype('int')
te['age'] = (abs(te['DAYS_BIRTH'])/365).astype('int')
```

```python
# days_employed

# 전처리
tr['DAYS_EMPLOYED'] = tr['DAYS_EMPLOYED'].apply(lambda x: 0 if x > 0 else x)
te['DAYS_EMPLOYED'] = te['DAYS_EMPLOYED'].apply(lambda x: 0 if x > 0 else x)

# 근무일수 -> 근무년차
tr['YEARS_EMPLOYED'] = (abs(tr['DAYS_EMPLOYED'])/365).astype('int')
te['YEARS_EMPLOYED'] = (abs(te['DAYS_EMPLOYED'])/365).astype('int')
```

```python
# begin_month

# 전처리
tr['begin_month'] = abs(tr['begin_month'])
te['begin_month'] = abs(te['begin_month'])

# 사용년수
tr['begin_years'] = (tr['begin_month']/12).astype('int')+1
te['begin_years'] = (te['begin_month']/12).astype('int')+1
```

```python
# income/가족수
tr['income_fam_ratio'] = tr['income_total']/tr['family_size']
te['income_fam_ratio'] = te['income_total']/te['family_size']
```

```python
# 일을 하게 된 시점(나이)
tr['start_employed'] = ((abs(tr['DAYS_BIRTH'])-abs(tr['DAYS_EMPLOYED']))/365).astype('int')
te['start_employed'] = ((abs(te['DAYS_BIRTH'])-abs(te['DAYS_EMPLOYED']))/365).astype('int')
```
