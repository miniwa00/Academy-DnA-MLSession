```python
import pandas as pd
```

```python
tr = pd.read_csv('train.csv', encoding = 'cp949')
te = pd.read_csv('test.csv', encoding = 'cp949')
```

```python
tr.isnull().sum()
```

```python
# 타겟값 형변환
tr['credit'] = tr['credit'].astype('str')
```

```python
# 결측치 처리
tr['occyp_type'] = tr['occyp_type'].fillna('no job')
te['occyp_type'] = te['occyp_type'].fillna('no job')
```

```python
# 값이 모두 같은 열 제거
tr.drop(columns = 'FLAG_MOBIL')
te.drop(columns = 'FLAG_MOBIL')
```

```python
# 출생일 -> 나이
tr['age'] = (abs(tr['DAYS_BIRTH'])/365).astype('int')
te['age'] = (abs(te['DAYS_BIRTH'])/365).astype('int')
```

```python
# 근무일수 -> 근무년차
tr['DAYS_EMPLOYED'] = tr['DAYS_EMPLOYED'].apply(lambda x: 0 if x > 0 else x)
te['DAYS_EMPLOYED'] = te['DAYS_EMPLOYED'].apply(lambda x: 0 if x > 0 else x)

tr['YEARS_EMPLOYED'] = (abs(tr['DAYS_EMPLOYED'])/365).astype('int')
te['YEARS_EMPLOYED'] = (abs(te['DAYS_EMPLOYED'])/365).astype('int')
```

```python
# 범주형 변수들 one hot
cat_features_tr = tr[['gender','car','reality','income_type','edu_type','family_type','house_type','occyp_type']]
cat_features_oh_tr = pd.get_dummies(cat_features_tr)
cat_features_oh_tr

cat_features_te = te[['gender','car','reality','income_type','edu_type','family_type','house_type','occyp_type']]
cat_features_oh_te = pd.get_dummies(cat_features_te)
cat_features_oh_te
```

```python
cat_features_oh_tr.to_csv('onehot_features_train.csv', index=False)
cat_features_oh_te.to_csv('onehot_features_test.csv', index=False)
```
